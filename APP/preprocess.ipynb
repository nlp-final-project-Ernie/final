```python
import jsonlines
import numpy as np
import json
import pandas as pd
```

```python
a = 0
feature=[]
text=[]
res=pd.DataFrame()
with open('data.jsonl', 'r', encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        # lst = list(data.values())
        #rand_num=a%len(lst[0]) #rand_num随机抽取特征关键词
        #print(a,len(lst[0]),rand_num,lst[0][0].sort())
        #feature.append(lst[0][0][1])
        #text.append(lst[4].strip("设计亮点"))
        dftest=pd.DataFrame(data)
        #print(dftest)
        res = pd.concat([res, dftest], ignore_index=True)
        a += 1
        if a ==1530 :
            break
```

```python
for i in range(res.shape[0]):
    # print(res.feature[i][0])
    if res.feature[i][0]!='版型' and res.feature[i][0]!='风格':
        res=res.drop(index=[i])
    else:
        res.feature[i]=res.feature[i][1]
        desc_to_remove='设计亮点\t\n'
        res.desc[i]=res.desc[i].replace('设计亮点','').replace('\t','').replace('\n','')
        #print(res.feature[i][1])
```
